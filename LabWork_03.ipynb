## Замена значений

заменим все значения 'female' на 'Woman', а 'male' – на 'Man'

```python
data1_copy = data1.copy()
data1_copy['Sex'] = data1_copy['Sex'].replace(['male','female'],['Man','Woman'])
data1_copy.head()
```
