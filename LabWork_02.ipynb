# Загрузка данных из внешних источников

Подключение библиотеки

```python
import pandas as pd
import numpy as np
```

## Загрузка данных из файла CSV

Получение готового набора данных с использованием `read_csv`

```python
url = 'https://raw.githubusercontent.com/enikolaev/AI_in_Professional_Area/main/data/work_02/data.csv'
data_csv = pd.read_csv(url)
data_csv.head()
```

Получение описания признаков набора данных

```python
data_csv.describe(include='all')
```

## Загрузка данных из файла Excel

Получение готового набора данных с использованием `read_excel`

```python
url = 'https://github.com/enikolaev/AI_in_Professional_Area/raw/main/data/work_02/data.xlsx'
data_excel = pd.read_excel(url,sheet_name=0)
data_excel.head()
```

Методы работы с полученным DataFrame не отличаются от аналогичного DataFrame, полученного с использованием `read_csv`

```python
data_excel.describe()
```

## Загрузка данных из файла JSON

Получение набора данных с использованием `read_json`

```python
url = 'https://raw.githubusercontent.com/enikolaev/AI_in_Professional_Area/main/data/work_02/data.json'
data_json = pd.read_json(url)
data_json.head()
```

```python
data_json.describe(include='all')
```
